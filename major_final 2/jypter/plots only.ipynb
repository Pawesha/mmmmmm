```python

# outlier na remove wala
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
from math import sqrt
# Load the preprocessed data
df = pd.read_excel("/Users/paweshashrestha/Downloads/major_final 2/cleaned files/refined_data_all.xlsx")
features = ['YEAR', 'MO', 'DY', 'HR', 'DOTW', 'IS_HOLIDAY?', 'T2M']
```

```python
# Create a separate box plot for each specified column
columns_names = ['electricity']
def generate_box_plots(df, columns_names):
  for column_name in columns_names:
    plt.figure(figsize=(8, 6))  # Adjust figure size
    df[[column_name]].boxplot()
    
    # Set y-axis ticks with an interval of 0.5
    plt.yticks([i for i in range(int(df[column_name].min()), int(df[column_name].max()) + 1, 1)])
    
    plt.title(f'Box Plot for Column {column_name}')
    plt.show()
```

```python
# Calculate the correlation with the 'electricity' column
correlations = df[features+ ['electricity']].corr()['electricity'].sort_values()

# Create a bar chart for the correlation values
plt.figure(figsize=(10, 6))
correlations.drop('electricity').plot(kind='bar')
plt.title('Correlation with Electricity Consumption')
plt.xlabel('Variables')
plt.ylabel('Correlation coefficient')
plt.axhline(y=0, color='black', linewidth=0.8)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

```python
# Group by 'YEAR' and sum the electricity consumption
total_electricity_by_year = df.groupby('YEAR')['electricity'].sum().reset_index()

# Plotting
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12, 18))

# Yearly electricity consumption bar graph
total_electricity_by_year.plot(kind='bar', x='YEAR', y='electricity', ax=axes, legend=False)
axes.set_title('Total Electricity Consumption by Year')
axes.set_xlabel('Year')
axes.set_ylabel('Total Electricity Consumption')
plt.show()
```

```python
# Group by 'DOTW' and calculate the mean electricity consumption
avg_electricity_by_dotw = df.groupby('DOTW')['electricity'].mean().reset_index()

# Plotting
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12, 18))

# Average electricity consumption by DOTW bar graph
avg_electricity_by_dotw.plot(kind='bar', x='DOTW', y='electricity', ax=axes, legend=False)
axes.set_title('Average Electricity Consumption by Day of the Week')
axes.set_xlabel('Day of the Week')
axes.set_ylabel('Average Electricity Consumption')

plt.show()
```

```python
# Group by 'HR' and calculate the mean electricity consumption
avg_electricity_by_hr = df.groupby('HR')['electricity'].mean().reset_index()

# Plotting
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12, 18))

# Average electricity consumption by HR bar graph
avg_electricity_by_hr.plot(kind='bar', x='HR', y='electricity', ax=axes, legend=False)
axes.set_title('Average Electricity Consumption by Hour of the Day')
axes.set_xlabel('Hour of the Day')
axes.set_ylabel('Average Electricity Consumption')

plt.show()
```
